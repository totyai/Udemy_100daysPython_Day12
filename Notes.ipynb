## Scope
- The variables could be either local or global
- Local scope variables only exists within function. 
    - It can only be accessed within the function it is declared, not outside.

Below, we get a name error, as the potion_strength doesn't make sense outside of *drink_potion* function.

```python
def drink_potion():
    potion_strength = 2
    print(potion_strength)

drink_potion()
print(potion_strength)
```

If we to have global scope, we need to declare it outside any fucntions.
Global variables can be called inside any function.

Not just variable but function is also wokring like that. These are called *namespace* -> where they are, defines the scope

```python
player_health = 10

def drink_potion():
    print(player_health)

drink_potion()
```

## Block Scope
**There is no block scope in Python**
This means that the local vs global scope differenciation is only applicable for function, *not* if statements. (like in other programing languages)
The below code would not work in other langs (*new_enemy* would not be reachable):

```python
game_level = 3
enemies = ["Skeleton", "Zombie", "Alien"]

if game_level < 5:
    new_enemy = enemies[0]

print(new_enemy)
```

If the condition is not met the *new_enemy* will not be created. Sometimes it is good to initialize the variables, to have a reserved name space, with a simple value.

## Exercise
**Prime Number Checker**
Prime numbers are numbers that can only be cleanly divided by themselves and 1. Wikipedia  
You need to write a function called is_prime() that checks whether if the number passed into it is a prime number or not.  It should return True or False.

e.g.
7 is a primer number because it is only divisible by 1 and itself.
But 4 is not a prime number because you can divide it by 1, 2 or 4.
NOTE: 2 is a prime number because it's only divisible by 1 and itself, but 1 is not a prime number because it is only divisible by 1.

Example Input 1
73

Example Output 1
True

Example Input 2
75

Example Output 2
False

```python
def is_prime(num):
    dividors = []
    for i in range(1, num+1):
        if (num % i) == 0:
            dividors.append(i)
    if len(dividors) == 2:
        return True
    else:
        return False

print(is_prime(73), is_prime(75))
```

## Modify Global Variables
If we use the same name for a variable inside a funciton, and have a global variable, then the complier will create two separate variables.
If we want to modify the global we need to declare with *global* key in local scope

```python
enemies = 1

def increase_enemies():
    global enemies
    enemies +=1
    print(f"enemies inside function: {enemies}")

increase_enemies()
```

## Python contants & global scope
In Python it is worth to use global scope for contants. Those that are never going to change.
There is a nameing convention for contants, to put the name in all UPPERCASE.

```python
PI = 3.1415

def my_func():
    print(PI)

my_func()
```
